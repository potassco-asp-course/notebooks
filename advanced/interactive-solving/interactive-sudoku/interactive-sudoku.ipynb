# Interactive Sudoku

The task of this project is the same as in the `interactive-solving`, 
but now the visualization uses `clingraph`:
* https://github.com/potassco/clingraph

Clingraph can be installed using conda:
```
conda install -c potassco clingraph
```

You can start from the script `interactive-sudoku.py` and follow the instructions 
in the comments.
